```python
import pandas as pd #in this line pandas acts as pd instead of pandas and its known as pandas package
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#dataset is a structure set of data and then pd.read_csv means it import a csv file to dataframe format
```

```python
dataset #dataset: it containt some data into it
```

```python
independent=dataset[["YearsExperience"]]#independent is a variable name its known for input variable and in this line we seperate the input column by using this variable name 
```

```python
independent
```

```python
dependent=dataset[["Salary"]] #dependent is a output variable , we split the input onput from this line and above lines
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) 
#in this line we split the train set and the test set sklearn is the library it cointain mechine learning process 
#in this process we can save memory
#we use some parameter into the function such as independent, test size,random state
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
# from using sklearn we import the LinearRegression
#regrossor is a variable for this model
#using "=" assigning symbol we assign the variable regressor to linear regression
#whenever we use the fit method to call a fuction it subtitute the values to the variable its just like a maths formulae
```

```python
weight=regressor.coef_
weight
#weight is a variable 
#Coef is nothing but its stores the weight calcution
#"."its known as dot operator its using for calling the object or function
```

```python
bais=regressor.intercept_
bais
#bais is a variable 
#like coef intercept stores the calcutlation of bais
```

```python
y_pred=regressor.predict(X_test)
#y_pred is a variable and predict is fuction of our application
# cross check process for our test data 
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
# in this line we find the r2 value using sklearn and then write a variable then assign a name for the variable 
#y_test is actual out put
#y_pred is predicted value of our model
```

```python
r_score
# if our models value is nearly to 1 its a good model 
```

```python
import pickle
filename="finalized_model_linear.sav"
# this is the process to save the model
#pickle is fuction 
```

```python
pickle.dump(regressor,open(filename,'wb'))#wb mean write binary
#using "pickle.dump" we dump the model into the filename
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))# rb mean read binary
result=loaded_model.predict([[14]])
#in this line we load the saved model
#we use the "loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))" this to load the model already saved model
#"loaded_model=" is a variable
```
